```python
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_CONFIG = "../dataset/data.yaml"

MODEL_NAME = "yolov8s-seg.pt"

EPOCHS = 100
IMG_SIZE = 1024
BATCH = 4
LR = 0.0005
PATIENCE = 30

print("Training config loaded.")
```

```python
model = YOLO(MODEL_NAME)
print("Model loaded:", MODEL_NAME)
```

```python
results = model.train(
    data=DATA_CONFIG,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH,
    lr0=LR,               
    patience=PATIENCE,    
    optimizer="AdamW",
    
    mosaic=1.0,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    perspective=0.0005,
    mixup=0.1,
    degrees=2,
    scale=0.3,
    shear=1.0,
)

results
```
